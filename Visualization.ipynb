### Visualization

In this workshop we will create basic visualizations with invented data. You will be able to plug in your own data to create similar plots and charts.

First, we import the Python packages. Numpy is for numerical tasks. Pandas is for working with data frames (tables). Matplotlib is a plotting package, and seaborn provides customization of matplotlib functions.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

We create some random data using Numpy functions.

```python

np.random.seed(42)
data = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C'], size=100),
    'Numeric1': np.random.randn(100),
    'Numeric2': np.random.randn(100),
})

data
```

### Scatter plot

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.scatter(data['Numeric1'], data['Numeric2'])
ax.set_title('Scatter Plot')
```

### Histogram

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.histplot(data['Numeric1'], bins=20, kde=True, color='skyblue')
ax.set_title('Histogram')
plt.show()
```

### Bar plot

```python
fig = plt.subplots(figsize=(8,6))
sns.countplot(x='Category', data=data,  palette='pastel', legend=False, hue = "Category")
ax.set_title('Bar Plot')
plt.show()
```

### Box plot

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.boxplot(x='Category', y='Numeric1', data=data, palette='husl', hue = "Category")
ax.set_title('Box Plot')
plt.show()
```

```python
# Example with binary data
binary_data = [0, 1, 1, 0, 1, 1, 0, 0, 1, 0]

# Create a count plot
sns.countplot(binary_data)
plt.title('Count Plot for Binary Data')
plt.show()
```

```python
# Example with binary data
counts = [binary_data.count(0), binary_data.count(1)]

# Create a pie chart
plt.pie(counts, labels=['0', '1'], autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart for Binary Data')
plt.show()
```

```python
# Example with binary data and groups
groups = ['Group A', 'Group B', 'Group C']
data = pd.DataFrame({'Group': np.random.choice(groups, size=100),
                     'Binary': np.random.choice([0, 1], size=100)})

# Create a stacked bar plot
pd.crosstab(data['Group'], data['Binary']).plot(kind='bar', stacked=True)
plt.title('Stacked Bar Plot for Binary Data by Group')
plt.show()
```

```python
# Example with a binary correlation matrix
binary_matrix = np.random.choice([0, 1], size=(10, 10))

# Create a heatmap
sns.heatmap(binary_matrix, cmap='Blues', annot=True, fmt='d')
plt.title('Binary Data Heatmap')
plt.show()
```
